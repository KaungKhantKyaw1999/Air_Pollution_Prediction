```python
import pandas as pd
import numpy as np
df = pd.read_csv('Predict.csv')
```

```python
df.head(100)
```

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df
```

```python
from datetime import datetime

df['Date'] = pd.to_datetime(df['Date'])
```

```python
print(df.loc[[37]])
```

```python
df.City = pd.Categorical(df.City)
c = df.City.astype('category')
print(dict(enumerate(c.cat.categories)))
df.City=df.City.cat.codes




df.Season = pd.Categorical(df.Season)
c = df.Season.astype('category')
print(dict(enumerate(c.cat.categories)))
df.Season=df.Season.cat.codes





df.Center = pd.Categorical(df.Center)
c = df.Center.astype('category')
print(dict(enumerate(c.cat.categories)))
df.Center=df.Center.cat.codes
```

```python
df.City.astype('category')
```

```python
df.dtypes
```

```python
y=df['PM2_5']
```

```python
X=df.drop(['PM2_5','PM1_0','PM10','AQI','Date','Year','New_cases','Cumulative_cases','New_deaths','Cumulative_deaths' ],axis=1)
```

```python
type(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1234)
X_train.shape, y_train.shape
X_train.shape, y_train.shape
```

```python
X.columns
```

```python
df
```

```python
is_NaN = df.isnull()
row_has_NaN = is_NaN.any(axis=1)
df[row_has_NaN]
```

```python
from xgboost.sklearn import XGBRegressor

#Fitting XGB regressor 
model = XGBRegressor(n_estimators=200)
```

```python
model.fit(X_train, y_train)
```

```python
model_score = model.score(X_test, y_test)
model_score
```

```python
 #Predict 
print(X_test)
model_pred = model.predict(X_test)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("Root mean squared error: %.2f"
      % sqrt(mean_squared_error(y_test, model_pred)))
```

```python
model.save_model('model.h5')
```

```python
import joblib
joblib.dump(model, 'Model.sav')
```
