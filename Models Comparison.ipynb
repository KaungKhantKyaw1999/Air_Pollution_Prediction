```python
import pandas as pd
import numpy as np
df = pd.read_csv('FinalData.csv')
```

```python
df.head(10)
```

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df.head(10)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = [16,9]

x = df['Month']
y = df['PM2_5']
plt.scatter(x,y)
plt.xlabel('Date')
plt.ylabel('PM2.5_ATM_ug/m3')
plt.title('PM2.5 by Month')
plt.show()
```

```python
from datetime import datetime

df['Date'] = pd.to_datetime(df['Date'])
```

```python

def time_series(start, end):
    time_series_df = df[['Date', 'PM2_5']][(df['Date'] >= start) & (df['Date'] <= end)]
    x = time_series_df.Date
    y = time_series_df.PM2_5
    plt.plot(x,y)
    plt.xlabel('Time')
    plt.ylabel('PM2.5 Value')
    plt.title('PM2.5 Time Series')
    return plt.show();
```

```python
time_series('2019','2021')
```

```python
time_series('2019-10','2019-11')
```

```python
df.dtypes
```

```python
df.City = pd.Categorical(df.City)
c = df.City.astype('category')
print(dict(enumerate(c.cat.categories)))
df.City=df.City.cat.codes




df.Season = pd.Categorical(df.Season)
c = df.Season.astype('category')
print(dict(enumerate(c.cat.categories)))
df.Season=df.Season.cat.codes





df.Center = pd.Categorical(df.Center)
c = df.Center.astype('category')
print(dict(enumerate(c.cat.categories)))
df.Center=df.Center.cat.codes
```

```python
df.dtypes
```

```python
y=df['PM2_5']
```

```python
X=df.drop(['PM2_5','PM1_0','PM10','AQI','Year','Date','New_cases', 'Cumulative_cases', 'New_deaths', 'New_deaths', 'Cumulative_deaths'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1234)
X_train.shape, y_train.shape
X_train.shape, y_train.shape
```

```python
X.columns
```

<h1>Linear Regression<h1>

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Create linear regression object
regr = LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
lin_pred = regr.predict(X_test)
```

```python
linear_regression_score = regr.score(X_test, y_test)
linear_regression_score
```

```python
from math import sqrt
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Root mean squared error: %.2f"
      % sqrt(mean_squared_error(y_test, lin_pred)))
# The absolute squared error
print("Mean absolute error: %.2f"
      % mean_absolute_error(y_test, lin_pred))
# Explained variance score: 1 is perfect prediction
print('R-squared: %.2f' % r2_score(y_test, lin_pred))
```

```python
plt.scatter(y_test, lin_pred)
plt.xlabel('Measured')
plt.ylabel('Predicted')
plt.title('Linear Regression Predicted vs Actual')
plt.show()
```

<h2>Neural Network</h2>

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Create MLPRegressor object
mlp = MLPRegressor()
```

```python
# Train the model using the training sets
mlp.fit(X_train, y_train)
```

```python
# Score the model
neural_network_regression_score = mlp.score(X_test, y_test)
neural_network_regression_score
```

```python
# Make predictions using the testing set
nnr_pred = mlp.predict(X_test)
```

```python
# The mean squared error
print("Root mean squared error: %.2f"
      % sqrt(mean_squared_error(y_test, nnr_pred)))
# The absolute squared error
print("Mean absolute error: %.2f"
      % mean_absolute_error(y_test, nnr_pred))
# Explained variance score: 1 is perfect prediction
print('R-squared: %.2f' % r2_score(y_test, nnr_pred))
```
